```python
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_data(filename,
              title,
              ylabel,
              legend_prefix,
              savefile):

    df = pd.read_csv(filename, header=None)


    plt.figure()
    plt.style.use("ggplot")
    plt.xlabel("Generation")
    plt.ylabel(ylabel)
    plt.title(title)
    for i in range(df.shape[1]):
        plt.plot(df[i], label=f"{legend_prefix} {i + 1}")
    plt.legend()
    plt.savefig(savefile)
```

```python
plot_data("fitnessData.csv",
          "Average Fitness of Creature Population",
          "Average Creature Fitness",
          "Simulation",
          "AverageFitness.png")
plot_data("minFitnessData.csv",
          "Minimum Fitness in Creature Population",
          "Minimum Creature Fitness",
          "Simulation",
          "MinFitness.png")
plot_data("populationSizeData.csv",
          "Population Size over Generations",
          "Population Size",
          "Simulation",
          "PopulationSize.png")
```
